```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5000)
classifier.fit(x_train,y_train)
n_pred=classifier.predict(x_test)
n_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,n_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,n_pred)
ac
```

```python
#svm
from sklearn.svm import SVC
classifier=SVC(kernel='linear')
classifier.fit(x_train,y_train)
s_pred=classifier.predict(x_test)
s_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,s_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,s_pred)
ac
```
